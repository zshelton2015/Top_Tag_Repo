```python
data=pd.read_csv('data/ML_data.csv',header=[0,1])
valid=pd.read_csv('data/Valid_data.csv')
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold

import matplotlib.pyplot as plt 
```

```python
data_dmatrix = xgb.DMatrix(data=data,label=valid)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(data, valid, test_size=0.2, random_state=1997)
X_train=xgb.DMatrix(X1_train)
X_test=xgb.DMatrix(X1_test)
y_train=xgb.DMatrix(y1_train)
y_test=xgb.DMatrix(y1_test)
```

```python
params={"objective":"binary:hinge",'num_parallel_tree':3,'colsample_bytree': 0.5,'learning_rate': .3,'early_stopping_round':60,
                'max_depth': 10, 'alpha':0}
xg_reg = xgb.train(params=params, dtrain=data_dmatrix, num_boost_round=500)
```

```python
xgb.plot_tree(xg_reg,num_trees=1)
plt.rcParams['figure.figsize'] = [100, 10]
plt.show()
```

```python
xgb.plot_importance(xg_reg,importance_type='cover')
```
